```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')

sns.pairplot(iris, hue='variety')
```

```python
iris.head()
```

```python
for col in iris.select_dtypes('float'):
    plt.figure()
    sns.distplot(iris[col])
```

```python
sns.clustermap(iris.corr(), annot=True)
```

```python
sns.jointplot('sepal.width', 'petal.length', data=iris, kind='kde', space=0, color='g')
```

```python
titanic = pd.read_excel('../titanic.xls')

sns.catplot(x='pclass', y='age', data=titanic, hue='survived', alpha=0.5)
```
