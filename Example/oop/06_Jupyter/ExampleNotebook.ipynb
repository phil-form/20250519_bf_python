```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,6))
plt.pie(df['survived'].value_counts(), labels=['died', 'survived'], autopct='%1.f%%')
plt.title('Title de mon graphique')
plt.show()
```

```python
df_survived = df[df['survived'] == 1]
df_died = df[df['survived'] == 0]

df['pclass'].value_counts()
```

```python
# Je r√©cup√®re les personnes qui ont plus de 30 ans
df[(df['age'] > 30)]
```

```python
# Je r√©cup√®re les personnes qui ont plus de 30 ans
df[(df['age'] > 30) & (df['survived'] == 1)]
```

```python
def age_cat(data):
    if(data['age'] < 18):
        return '-18'
    elif(data['age'] < 30):
        return '20s'
    elif(data['age'] < 60):
        return '30-60'
    else:
        return '60+'

df['age_cat'] = df.apply(age_cat, axis=1)
df.head()
```

```python
df.select_dtypes('float')
```

```python
df = df.drop(columns=['body', 'fare'])

for col in df.select_dtypes('float'):
    plt.figure(figsize=(20,10))
    sns.displot(df_survived[col], label='survived')
    sns.displot(df_died[col], label='died')
    plt.legend()
```

```python
for col in df.select_dtypes('float'):
    plt.figure(figsize=(20,10))
    sns.displot(data=df, x=col, hue='survived')
    plt.legend()
```
