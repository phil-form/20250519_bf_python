## Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 4, 10)
y = x ** 2

plt.plot(x, y)
plt.show() # show the graph (if in a loop or outside of notebooks)
```

## Most common problems :
- different shape between x and y. Both must have the same dimensions.

```python
x = np.linspace(0, 4, 10)
y = np.linspace(0, 4, 5)

plt.plot(x, y)
plt.show() # show the graph (if in a loop or outside of notebooks)
```

```python
x = np.linspace(0, 4, 10)
y = x ** 2

plt.scatter(x, y)
plt.show() # show the graph (if in a loop or outside of notebooks)
```

# Parameters 

## Color

```python
x = np.linspace(0, 4, 10)
y = x ** 2

plt.plot(x, y, c='r')
plt.show() # show the graph (if in a loop or outside of notebooks)
```

## Line width

```python
x = np.linspace(0, 4, 10)
y = x ** 2

plt.plot(x, y, lw=5)
plt.show() # show the graph (if in a loop or outside of notebooks)
```

## Line style

```python
x = np.linspace(0, 4, 10)
y = x ** 2

plt.plot(x, y, ls="--")
plt.show() # show the graph (if in a loop or outside of notebooks)
```

# Good practice

```python
plt.figure(figsize=(12, 8)) # first declare that we want to draw a figure
                            # figsize are in inch
plt.plot(x, y) # now we fill the figure with datas
```

```python
plt.figure(figsize=(12, 8)) # first declare that we want to draw a figure
                            # figsize are in inch
plt.plot(x, y, label='quadratic') # now we fill the figure with datas
plt.plot(x, x**3, label='cubic') # now we fill the figure with datas
plt.plot(x, np.sin(x), label='sin') # now we fill the figure with datas
plt.title('example 1')
plt.xlabel('x name')
plt.ylabel('y name')
plt.legend()
plt.show()

plt.savefig('test.png')

plt.figure(figsize=(12, 8)) # first declare that we want to draw a figure
                            # figsize are in inch
#plt.plot(x, y, label='quadratic') # now we fill the figure with datas
#plt.plot(x, x**3, label='cubic') # now we fill the figure with datas
plt.plot(x, np.sin(x), label='sin') # now we fill the figure with datas
plt.title('example 1')
plt.xlabel('x name')
plt.ylabel('y name')
plt.legend()
plt.show()

plt.savefig('test.png')
```

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)}
```

```python
def graphic(data):
    n = len(data)
    plt.figure(figsize=(12, 20))
    
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
        
    plt.show()
```

```python
graphic(dataset)
```
