# Most useful graphics in matplotlib

## 5) Scatter

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target

names = list(iris.target_names)

print(names)
print(x.shape)
print(np.unique(y).size) # 3 different type
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5)

plt.subplot(2, 2, 2)
plt.scatter(x[:, 1], x[:, 2], c=y, alpha=0.5)

plt.subplot(2, 2, 3)
plt.scatter(x[:, 2], x[:, 3], c=y, alpha=0.5)

plt.subplot(2, 2, 4)
plt.scatter(x[:, 3], x[:, 0], c=y, alpha=0.5)
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2]*50)
```

# 3) Histplot

```python

x = np.random.randn(1000)

plt.figure(figsize=(12, 10))
#plt.subplot(121)
plt.hist(x, bins=10)
plt.title('bins=10')
#plt.subplot(122)
plt.hist(x, bins=50)
plt.title('bins= 50')
plt.show()
```

```python
x = iris.data

plt.hist2d(x[:,0], x[:,1], cmap='Blues')
plt.xlabel('longueur spal')
plt.ylabel('largeur spal')
plt.colorbar()
```

```python

# histogramme d'une image
from scipy import misc
face = misc.face(gray=True)

plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.imshow(face, cmap='gray')
plt.subplot(122)
plt.hist(face.ravel(), bins=255)
plt.show()
```

# 2) ContourPlot

```python

f = lambda x, y: np.sin(x) + np.cos(x+y)*np.cos(x)

X = np.linspace(0, 5, 50)
Y = np.linspace(0, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

plt.contour(X, Y, Z, 20, colors='black') # able us to visualize data from the top view.
                                         # this can help to understand data and to analyse or optimize a function
```

```python

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

# 1) imshow

```python

plt.figure(figsize=(12, 3))

# Simple graphique imshow()
X = np.random.randn(50, 50)

plt.subplot(131)
plt.imshow(X)
plt.title('random normal')

# Matrice de corrlation des iris
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target


plt.subplot(132)
plt.imshow(np.corrcoef(X.T, y)) # transpose because we want to know the correlation between columns and not rows
plt.title('Iris Corrlation')

# Matrice f(X, Y) = sin(X) + cos(Y)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)

plt.subplot(133)
plt.imshow(f(X, Y))
plt.colorbar()
plt.title('f(x, y) = sin(x) + cos(y)')
```
