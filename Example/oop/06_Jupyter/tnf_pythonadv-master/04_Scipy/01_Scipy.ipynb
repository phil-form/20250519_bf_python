# Interpolate

We can have two sensors that collect data at a different rate.

Thus we will have NaNs on one of them.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
x = np.linspace(0, 10, 10)
y = x ** 2
plt.scatter(x, y)
```

```python
f = interp1d(x, y, kind='linear')

new_x = np.linspace(0, 10, 30)
result = f(new_x)

plt.scatter(x, y)
plt.scatter(new_x, result, c='r')
```

```python
x = np.linspace(0, 10, 10)
x2 = np.linspace
y = np.sin(x)
plt.scatter(x, y)
```

interp1d will return a function that we will be able to use to interpolate.



```python

# création de la fonction interpolation f
f = interp1d(x, y, kind='cubic')

# résultats de la fonction interpolation f sur de nouvelles données
new_x = np.linspace(0, 10, 50)
result = f(new_x)

# visualisation avec matplotlib
plt.scatter(x, y)
plt.plot(new_x, result, c='r')
```

# Optimize

```python
x = np.linspace(0, 2, 100)
y = 1/3*x**3 - 3/5 * x**2 + 2 + np.random.randn(x.shape[0])/20
plt.scatter(x, y) 
```

```python
def f (x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d
```

```python
from scipy import optimize
```

```python
params, param_cov = optimize.curve_fit(f, x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]), c='g', lw=3)
```

```python
def f2(x):
    return x ** 2 + 10 * np.sin(x)
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, f2(x))
```

```python
optimize.minimize(f2, x0=-8)
```
